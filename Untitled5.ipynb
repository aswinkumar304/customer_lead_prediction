```python
dataset.isna().sum()
```

```python
for col in dataset.columns:
    if dataset[col].dtypes == 'int64' or dataset[col].dtypes == 'float64':
        dataset[col].fillna(dataset[col].mean(), inplace=True)
    else:
        dataset[col].fillna(dataset[col].mode()[0], inplace=True)
```

```python
# we drop the value with 1 occurence
dataset.drop(dataset.index[dataset['Lead Origin'] == 'Quick Add Form'], inplace=True)
# we plot the value counts with respect to the target variable
fig, axs = plt.subplots(figsize = (15,7.5))
sns.countplot(x = "Lead Origin", hue = "Converted", data = dataset, order = dataset['Lead Origin'].value_counts().index)
xticks(rotation = 45)
plt.show()
```

```python
# here we calculate the conversion ratio for each value in Lead Origin
d = {}
for val in dataset['Lead Origin'].unique():
    a = dataset[dataset['Lead Origin'] == val]['Converted'].sum()
    b = dataset[dataset['Lead Origin']==val]['Converted'].count()
    d[val] = [a, b, round(a/b*100, 2)]
pd.DataFrame.from_dict(d, orient='index').rename(columns = {0: 'Converted', 1: 'Leads',2: 'Conversion Ratio'}).sort_values(by=['Conversion Ratio'], ascending=False)
```
